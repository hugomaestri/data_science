Covid-19, caused by the SARS-CoV-2 virus, is a contagious respiratory illness that emerged as a global pandemic in late 2019. The disease is part of a larger family of coronaviruses that can cause illnesses ranging from the common cold to more severe diseases like Middle East Respiratory Syndrome (MERS) and Severe Acute Respiratory Syndrome (SARS).
Covid-19 was first identified in December 2019 in the city of Wuhan, Hubei province, China. Initially linked to a seafood market, the virus quickly spread, and by March 2020, the World Health Organization (WHO) declared it a pandemic. The virus primarily spreads through respiratory droplets when an infected person coughs, sneezes, or talks. It can also spread by touching surfaces contaminated with the virus and then touching the face. Due to its high transmissibility, Covid-19 rapidly disseminated across the globe, affecting millions of people in a matter of months.

The Covid-19 pandemic has had a profound impact on the global economy. Lockdowns, travel restrictions, and the closure of businesses led to a decline in economic activity, triggering recessions in many countries. Industries such as tourism, hospitality, and retail were particularly hard hit, leading to mass unemployment and financial instability for millions of people. Governments worldwide implemented stimulus packages and financial aid to mitigate the economic fallout, but the long-term effects of the pandemic continue to reshape global economic landscapes.

In this notebook, various Covid-19 data including confirmed cases, deaths, hospitalizations, testing, and vaccination, will be analyzed to understand how the pandemic developed.

### Data Acquisition
The dataset has been developed by [*Our World in Data*](https://ourworldindata.org/) whose goal is to make the knowledge on the big problems accessible and understandable. Our World in Data (OWID) has collected COVID-19 data from various domains since the pandemic started. The dataset includes geographical and historical data on the pandemic, such as confirmed cases, deaths, hospitalizations, testing, vaccination, age groups, pre-existing conditions, and more.

```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

# dataset path
DATA_PATH = "datasets\\owid-covid-data.csv"

# import the dataset to a dataframe with column date as the index
df = pd.read_csv(DATA_PATH)


# first 5 inputs of the dataset
df.head()
```

#### 2 - Dataset size


```python
print("Dataset size:",
      "\nNumber of entries:", df.shape[0],
      "\nNumber of variables:", df.shape[1],
      "\n   ----------------------   ")

# a more general approach to display those information
df.info()
```

The dataset contains nearly 430,000 entries and 67 variables.

```python
# convert date column to datetime format
df["date"] = pd.to_datetime(df["date"])
```

#### 3 - Null Values

```python
# Set the maximum number of rows to be displayed
pd.set_option("display.max_rows", None)

print("Percentage of null values by columns:")
round((df.isnull().sum() / df.shape[0]) * 100, 2).sort_values(ascending=False)
```

As seen above, some variables contains a significant amount of null values. The main goal here is to analyze the number of cases, deaths, vaccinations, location and date. For these key variables, the null values represents:
- `total_cases` -> 4.11% of null values
- `total_deaths` -> 4.11% of null values
- `total_vaccinations` -> 80.11% of null values
- `location` and `date` -> 0.00% of null values 

Itâ€™s important to note that not all of these null values indicate missing data; in some cases, they represent actual information where there were zero occurrences. For example, on the early stages of the pandemic, there were no confirmed cases, meaning the total cases were still zero.  In this context, the 'null value' represented by this zero actually indicates that there were no cases reported yet.

#### 4 - Confirmed cases and deaths per location

```python
# Check all locations in the dataset
df["location"].value_counts().sort_index(ascending=True)
```

```python
df_clean = df.copy()
```

```python
# remove location that are not a country
df_clean = df_clean.drop(df_clean[df_clean["location"] == "High-income countries"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "European Union (27)"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Upper-middle-income countries"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Lower-middle-income countries"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Low-income countries"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Asia"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "World"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Europe"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Oceania"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "Africa"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "North America"].index)
df_clean = df_clean.drop(df_clean[df_clean["location"] == "South America"].index)
```

```python
# look for the most recent date with confirmed cases
most_recent_date = df_clean.index.min()
for i in df_clean.index:
    if df["total_cases"][i] != 0:
        most_recent_date = df_clean["date"][i]

print(f"The most recent date with a confirmed case is", most_recent_date.date())
```

```python
total_cases = df_clean.loc[df_clean["date"] == most_recent_date, ["location", "total_cases"]].sort_values(by="total_cases", ascending=False)
total_cases_scaled = total_cases.copy()
total_cases_scaled["total_cases"] = total_cases_scaled["total_cases"]/1000000

fig, ax = plt.subplots()
total_cases_scaled[0:10].plot(x="location", y="total_cases", kind="bar", ax=ax)
ax.set_ylabel("Total Cases [millions]")
ax.set_xlabel("")
ax.set_title("Top 10 Countries in Number of cases")

plt.tight_layout()
```

```python
total_deaths = df_clean.loc[df_clean["date"] == most_recent_date, ["location", "total_deaths"]].sort_values(by="total_deaths", ascending=False)
total_deaths_scaled = total_deaths.copy()
total_deaths_scaled["total_deaths"] = total_deaths_scaled["total_deaths"]/1000

fig, ax = plt.subplots()
total_deaths_scaled[0:10].plot(x="location", y="total_deaths", kind="bar", ax=ax)
ax.set_ylabel("Total deaths [thousands]")
ax.set_xlabel("")
ax.set_title("Top 10 Countries in Number of Covid-19 Deaths")

plt.tight_layout()
```
